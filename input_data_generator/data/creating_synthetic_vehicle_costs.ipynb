```python
import pandas as pd
```

```python
df = pd.read_csv('synthetic_travel_times.csv')
```

```python
df.head()
```

```python
output_list = [] 
for index, row in df.iterrows():
    base_cost = row.travel_time_ms / 3600000.0
    for required_capacity in range(100,2100,100):
        cost = base_cost * (100.0 + (required_capacity-100)/4)
        vehicles =  "100.0:" + str(int(required_capacity / 100))
        output_list.append((row.orig_gh, row.dest_gh, required_capacity, cost, vehicles))
```

```python
df_output = pd.DataFrame(output_list, columns = ['orig_gh', 'dest_gh', 'required_capacity', 'cost', 'vehicles'])
df_output.to_csv('synthetic_vehicle_costs.csv', index = False)
```
